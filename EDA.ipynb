```python
# Encontramos que algunass filas tienen la duracion en la columna de rating, buscamos corregir

mask = df['rating'].str.find('min')>0 # Creamos una mascara en la que es verdadero si encuentra la palabra min en la columna rating

df.loc[mask, 'duration'] = df[df['rating'].str.find('min')>0].rating # Usamos la mascara pasar en esas columnas la informacion de rating a duracion
df.loc[mask, 'rating'] = pd.NA # En esas mismas filas, un cambio para remover la informaci√≥n de rating que debia estar en duraci√≥n

# Usamos el mismo procedimiento pero para las que contengan season, tambien esta seasons, pero si contiene season, contiene Seasons

mask = df['rating'].str.find('Season')>0
df.loc[mask, 'duration'] = df[df['rating'].str.find('min')>0].rating # Usamos la mascara pasar en esas columnas la informacion de rating a duracion
df.loc[mask, 'rating'] = pd.NA


# Verificamos con un ejemplo que se debio corregir
df[df['title'] == 'Loro']
```

Ahora, buscamos y llenamos la informacion que tenemos de otras plataformas o de la misma pelicula en otro idioma

Encontramos las peliculas/ series que estan repetidas, teniendo el mismo titulo o la misma descripci√≥n. La descripcion porque algunos titulos son diferentes - EJ: Coco y Coco(In Spanish)- pero que, por la descripcion, podemos saber que es la misma pelicula/serie.


```python
# Encontramos los repetidos de titulo y descripcion
duplicates = pd.concat([df[df.title.duplicated(keep = False)], df[df.description.duplicated(keep = False)]])

# Quitamos duplicados, los que tenian tanto duplicado de nombre como descripci√≥n
duplicates.drop_duplicates(inplace=True)

# Tomamos todas las columnas
columns = duplicates.columns

columns = columns.drop(labels= ['date_added', 'type', 'title','platform'] ) # Quitamos sobre las cuales no podemos obtener informaci√≥n de otro lado

empties = duplicates.isnull() # Tenemos en cuenta solo las filas que tienen nulos, es decir, que les falta informaci√≥n

#Recorremos los valores duplicados, manteniendo el indice original
for ind in duplicates.index: # Recorremos filas por indica
    
    for col in columns: # recorremos las columnas
        
        if (empties.loc[ind, col]): # Encontramos si el valor  especifico es nulo y queremos buscar mas informacion para remplazar
    
            # Encontramos las otras filas que tienen el mismo titulo y son del mismo tipo. Esto ultimo para asegurar que son la misma 
            # Serie/Pelicula y no una serie que se llama igual a la pelicula y viceversa
            row1 = duplicates.mask(((duplicates['title'] == duplicates['title'][ind]) & (duplicates['type'] == duplicates['type'][ind])) & (duplicates.index != ind))
            # Encontramos las otras filas que tienen la misma descripcion y tambien son del mismo tipo
            row2 = duplicates[(duplicates['description'] == duplicates['description'][ind]) & (duplicates['type'] == duplicates['type'][ind]) & (duplicates.index != ind)]
            
            rows = pd.concat([row1, row2]) # Concatenamos ambos dataframes para usarlos como uno solo
            rows.drop_duplicates(inplace = True) # removemos duplicados exactos para recorrer menos filas
            
            for i in rows.index: # Recorremos el subset de los duplicados de la fila en cuestion
                if(rows.isnull().loc[i, col] == False): # buscamos en solo la columna en cuestion
                    # Si encontramos el valor no nulo de una columna, remplazamos el nulo por ese
                    df.loc[ind,col] = rows.loc[i, col]
                    break #Como solo tenemos una fila especifica y una columna especifica, una vez se remplaza, no necesita seguir buscando mas
                
# al final debemos tener toda la informacion que tenemos de todos los duplicados, ya aplicada
```

Ahora queremos una tabla de los actores que hacen parte de los show y peliculas, y una tabla que nos relaciona los actores con los shows y peliculas en los que aparecen (junction table). Esta celda demora unos 30 minutos en completarse en mi computador

```python
# Creamos listas que despues convertiremos en dataframes
actors = []
junctionActors = []
nules = df['cast'].isnull()
# Recorremos todas las filas de el dataframe conbinado
for ind in df.index:
          # Creamos una lista de los actores desde la columna cast, separando por comas, para cada show o pelicula
    lista = str(df['cast'][ind]).split(",")
   
    # Consideramos si la lista esta vacia
    if(len(lista)>1):
    
        for actor in lista:# Recorremos esta lista de actores del show o pelicula especifico
            # Vamos a usar un algorimo de comparacion de palabras por la distancia de levenshtein.
            # En este caso se va a comparar la lista de actores que se va creando con los que vamos verificando y determinando
            # que probabilidad hay que sea el mismo que ya esta en la lista
            res = process.extractOne(actor, actors, scorer=fuzz.token_ratio, score_cutoff= 98)
            #comparamos con el resto de la lista de actores, con dos metodos y tomando el mejor. 
            # Para mas informacion visite https://maxbachmann.github.io/RapidFuzz/Usage/fuzz.html
            score = 0 # Inicializamos el score
            if(res!=None):
                cat, score, pos = res #desempacamos la tupla, si se encontraron valores
                print(actor, cat, score, pos) # imprimimos cada comparaci√≥n, por si queremos hacer verificaciones a mano
            if(score > 98): #
                # si el actor ya esta en la lista de actores, solo agregamos la relacion a junction table
                junctionActors.append({'idShow': ind, 'idActor': actor.index(actor)}) #pos
            else:
                # Si no esta en la lista, lo agregamos y agregamos la relacion en la junction table
                actors.append(actor)
                junctionActors.append({'idShow': ind, 'idActor': len(actors)-1})

# No hay problema en reducir esta lista pues el usuario no va a introducir el nombre del actor, sino que dados otros parametros
# el resultado ser√° el nombre del autor. 
# Note que la junction table usa como primary key de los actores, el indice que tiene en la tabla de actores.
# En total este metodo reduce la lista de actores en unos 4000 valores, frente a una comparacion directa

#Vemos la lista de actores
display(actors)
```

Procedemos tambien con las categorias de la misma forma, tambien creando una junction table

```python
#Creamos listas que despues convertiremos en dataframes

categories =[]
junctionCategories = []

for ind in df.index:
    #Creamos una lista de las categorias desde la columna listed_in, separando por comas
    lista = str(df['listed_in'][ind]).split(",")
    #recorremos esta lista de categorias
    for category in lista:
        category =  (category.strip()) # removemos espacios en blanco y otros problemas del nombre.
        if(category in categories):
            #si la categoria ya esta en la lista de categorias, solo agregamos la relacion a junction table
            junctionCategories.append({'idShow': ind, 'idCategory': categories.index(category) })
        else:
            # Si no esta en la lista, lo agregamos y agregamos la relacion en la junction table
            categories.append(category)
            junctionCategories.append({'idShow': ind, 'idCategory': categories.index(category) })

# dado que las categorias si las va a indicar el usuario en los request, las dejamos sin alterar ni tratar de reducir


display(categories) # visualizamos la lista
    

```

Convertimos las listas creadas en dataframes
